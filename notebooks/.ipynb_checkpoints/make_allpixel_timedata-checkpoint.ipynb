# Make Time-Series Data for all Pixels
authors: Tuguldur Sukhbold

This script creates time-series data for each pixel in a sample Sentinel-2 image. This data will be used to develop a model for classifying each pixel into seasonal vs non-seasonal classes based on their time variability. As a sample case we will be extracting data from filters #4 and #13 for fieldID=ff4174be.

We will be using two functions developed in the sentinel_timeseries notebook; so let's load them and first visualize the data:

```python
datapath = '../d/sentinel/'


import numpy as np
import matplotlib.pyplot as plt
from sys import exit
from itertools import product as iterp
from skimage.io import imread as rdtif


def getSentinelTimeData(path2tif, filterCloudy=True):
    """
    process input Sentinel TIFF file to map its contents into a dictionary
    with filter keys and monthly images as values
    
    note: it won't process images with number of channels other than 192
    
    when filterCloudy flag is True:
    - excludes cloudy images based on cloud-mask (filter#16)
    - final dictionary contains 13 real filter data
    """
    
    # filter and month IDs
    filters = np.arange(16, dtype=int) + 1
    months  = np.arange(12) + 1

    # read TIFF file into 3D array
    img = rdtif(path2tif)
    
    # stop if its a funky image, proceed only if there are usual 192 channels:
    if img.shape[-1] != 192: exit(f' cannot process this funky image with {img.shape[-1]} channels')
    
    # initialize the dict with empty list for each filter
    d = {}
    for f in filters: d[f] = []
    
    # populate with 2D images
    for i, j in iterp(months, filters):

        channel = (i - 1) * len(filters) + j - 1
        
        # append normalized image
        maxFrame = np.amax(img[:, :, channel])
        if maxFrame == 0.:
            d[j].append(img[:, :, channel])
        else:
            d[j].append(img[:, :, channel] / maxFrame)
    
    # exclude cloudy images
    if filterCloudy:
        
        for f in filters:
            for month in months:
                
                # max value of cloud mask image
                maxCloudMask = np.amax(d[16][month-1])
                
                # its cloudy if max is not 0
                if maxCloudMask != 0: d[f][month-1] = None
        
        # we don't need the last 3 elements
        del d[16] # cloudmask itself
        del d[15] # QA20
        del d[14] # QA10

    return d


def getPixelIntensity(x, y, img, cyclic=True):
    """
    returns intensity as a function of time
    for a given pixel (x,y) in given list of 
    2D image arrays of 12 months
    
    when cyclic flag is True:
    - periodic linear interpolation to fill missing data
    """
    
    months = np.arange(len(img)) + 1
    
    t = [m             for m in months if img[m-1] is not None]
    i = [img[m-1][x,y] for m in months if img[m-1] is not None]
    
    # interpolate to fill cloudy data
    if cyclic:
        months = np.arange(12) + 1
        i = np.interp(months, t, i, period=len(months))
        t = months
    
    return np.array(t), np.array(i)


# field ID and filter ID
fieldID  = 'ff4174be'
filterID = 4

# TIFF image filename
fname = f'{fieldID}.tif'

# load all data
d = getSentinelTimeData(f'{datapath}{fname}')

# image from sample channel, can be any of them
# only used to set the loop limits and to get total num of pixels
sampleChannel = d[filterID][4]
allpx = sampleChannel.shape[0] * sampleChannel.shape[-1]


fig, ax = plt.subplots(figsize=(12,4))

for i,j in np.ndindex(sampleChannel.shape):
    x,y = getPixelIntensity(i, j, d[filterID])
    ax.plot(x, y, color='gray', lw=0.1, alpha=0.5)

ax.set(xlabel='month', ylabel='normalized intensity', title=f'time-variation of all {allpx} pixels of filter#{filterID}')
plt.show()
```

Notice the sinusoidal cluster of curves - these are the seasonal pixels that we want to separate from the rest. Now let's write the data:

```python
def writePixelTimeData(d, filterID):
    """
    writes time series data for all pixels of a given image filter
    
    d = SentinelTimeData
    filterID = filter number
    """

    # image from sample channel, can be any of them
    # only used to set the loop limits and to get total num of pixels
    sampleChannel = d[filterID][4]
    allpx = sampleChannel.shape[0] * sampleChannel.shape[-1]
    
    # saved data filename
    savename = f'../data/px_data_field_{fieldID}_filter_{filterID}.dat'

    
    def write_header(df):
        # function that writes data header
        df.write('#\n')
        df.write(f'# normalized intensity time variation of all {allpx} pixels for:\n')
        df.write(f'# fieldID  = {fieldID}\n')
        df.write(f'# filterID = {filterID}\n')
        df.write(f'# datapath = {datapath}\n')
        df.write('#\n')
        df.write('# x  y  i[month#1]  i[month#2]   i[month#3]   ...   i[month#12]\n')

        
    with open(savename, 'w') as df:
        write_header(df)
        for i,j in np.ndindex(sampleChannel.shape):
            x,y = getPixelIntensity(i, j, d[filterID])
            line = f'{i:02d}   {j:02d}'
            for month in x: line += f'   {y[month-1]:.4f}'
            df.write(f'{line}\n')
    
    print(f' data written to {savename}')

    
# write for filters 4 and 13
writePixelTimeData(d, 4)
writePixelTimeData(d,13)
```

The data files are written in ../data/ directory. Filenames indicate which field and filter combination, and the same info are also written in file header. These are simple ascii tables and can be read in many different ways; here is one example:

```python
# let's read everything into a dictionary where the tuple of (x,y) 
# coordinates are keys, and intensity variations as values
pixel = {}

fname = '../data/px_data_field_ff4174be_filter_4.dat'
with open(fname, 'r') as file:
    # read all lines
    lines = file.readlines()
    
    # all header info into header key
    header = [line.strip() for line in lines[:7]]
    pixel['header'] = header
    lines = lines[7:]
    
    # main data
    for line in lines:
        xy = tuple([int(val) for val in line.strip().split()[0:2]])
        intensity = np.array([float(val) for val in line.strip().split()[2:]])
        pixel[xy] = intensity

# data for x=0 and y=0
pixel[(0,0)]
```

As a check let's recreate the plot we made above using our dictionary:

```python
fig, ax = plt.subplots(figsize=(12,4))

for i,j in np.ndindex(sampleChannel.shape):
    x,y = np.arange(12)+1, pixel[(i,j)]
    ax.plot(x, y, color='gray', lw=0.1, alpha=0.5)

ax.set(xlabel='month', ylabel='normalized intensity', title=f'time-variation of all {allpx} pixels of filter#{filterID}')
plt.show()
```
